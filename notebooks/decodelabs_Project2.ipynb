# Project 2: Exploratory Data Analysis (EDA)

```python
# 1. Calculate basic statistics (mean, median, min, max for numerical columns)
print("--- Basic Statistics ---")
print(df[['Quantity', 'UnitPrice', 'TotalPrice']].describe())
```

```python
# 2. Identify trends (e.g., Total Sales per Product)
product_sales = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)
print("\n--- Top Selling Products ---")
print(product_sales)
```

```python
# 3. Spot Outliers (Looking at extreme Total Price values)
# You can use standard deviation or IQR. Let's look at the max values:
outliers = df[df['TotalPrice'] > df['TotalPrice'].quantile(0.99)]
print("\n--- Top 1% Outlier Transactions ---")
print(outliers[['OrderID', 'Product', 'TotalPrice']])
```
