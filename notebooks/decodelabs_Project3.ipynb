# Project 3: SQL Data Analysis

```python
import sqlite3
import pandas as pd
```

```python
# 1. Loaded my cleaned dataset from Project 1
df = pd.read_csv("C:/Users/User/Desktop/DecodeLabs DA Internship/Cleaned_Sales_Data.csv")
```

```python
# 2. Created a connection to a local SQL database 
conn = sqlite3.connect('ecommerce.db')
```

```python
# 3. Pushing the pandas DataFrame into a SQL table named 'sales_data'
df.to_sql('sales_data', conn, if_exists='replace', index=False)
```

```python
print("Database connected and table 'sales_data' created successfully!")
```

```python
query1 = """
SELECT OrderID, Date, Product, TotalPrice 
FROM sales_data 
WHERE OrderStatus = 'Delivered' AND TotalPrice > 1000
ORDER BY TotalPrice DESC;
"""

# Executing the query against my database connection
high_value_deliveries = pd.read_sql_query(query1, conn)

# Display the results
high_value_deliveries
```

```python
query2 = """
SELECT 
    Product, 
    COUNT(OrderID) AS Total_Orders, 
    ROUND(AVG(TotalPrice), 2) AS Avg_Order_Value, 
    ROUND(SUM(TotalPrice), 2) AS Total_Revenue
FROM sales_data
GROUP BY Product
ORDER BY Total_Revenue DESC;
"""

# Executing the query
product_summary = pd.read_sql_query(query2, conn)

# Displaying the results
product_summary
```

```python
# Closing the connection so i can free up memory
conn.close()
```
