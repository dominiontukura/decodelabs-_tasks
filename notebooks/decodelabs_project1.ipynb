# Project 1: Data Cleaning & Preparation

```python
import pandas as pd

# 1. Load the raw dataset
df = pd.read_excel("C:/Users/User/Desktop/DecodeLabs DA Internship/Dataset for Data Analytics_RAW.xlsx")
```

```python
# 2. Identify missing or null values
print(df.isnull().sum()) # Shows missing values per column
```

```python
# 3. Handle missing values (Fill empty Coupon Codes with 'NONE')
df['CouponCode'] = df['CouponCode'].fillna('NONE')
```

```python
# 4. Remove duplicates (if any existed)
df = df.drop_duplicates()
```

```python
# 5. Correct data formats (Convert 'Date' from Text to Datetime)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Save the cleaned data to a new file
df.to_csv("Cleaned_Sales_Data.csv", index=False)
```
